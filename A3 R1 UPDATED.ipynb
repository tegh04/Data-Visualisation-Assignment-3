```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
imports = pd.read_excel("A3 data.xlsx", sheet_name=0)
exports = pd.read_excel("A3 data.xlsx", sheet_name=1)
```

```python
exports.drop(['Unnamed: 79', 'Unnamed: 80'], axis=1, inplace=True)
```

```python
imports.drop([0, 1], inplace=True)
exports.drop([0, 1], inplace=True)
```

```python
exports.loc[:, 'Export Category'] = exports['Export Category'].str.replace('Y', '')
imports.loc[:, 'Import Category'] = imports['Import Category'].str.replace('Y', '')
```

```python
imports.set_index('Import Category', inplace=True)
exports.set_index('Export Category', inplace=True)
```

```python
data = pd.concat([imports, exports], keys = ['imports', 'exports'], axis=1)
```

```python
data
```

```python
data = data.astype('int64')
```

```python
#Get Diff
diff = data['exports'][:] - data['imports'][:] 
```

```python
diff
```

```python
#Fix data up after fixing difference
data.columns = data.columns.to_flat_index()
```

```python
data
```

```python
# Get ratio data
shape = data.shape
ratio_data = np.tile(1.0, shape)
for i in range(1, 33):
    ratio_data[i][:] = data.iloc[i, :]/data.iloc[i-1, :]
```

```python
ratio = pd.DataFrame(ratio_data, index = data.index, columns = data.columns)
```

```python
ratio
```

```python
# Get stats
stats = data.describe()
```

```python
# pct: percentages 
pct = data.copy()

for col in pct.columns:
        
    if col[1].startswith('0 '):
        divisor = pct[[col]]
            
    elif col[1].startswith('1 '):
        divisor = pct[[col]]
        
    elif col[1].startswith('2 '):
        divisor = pct[[col]]
            
    elif col[1].startswith('3 '):
        divisor = pct[[col]]
        
    elif col[1].startswith('4 '):
        divisor = pct[[col]]
            
    elif col[1].startswith('5 '):
        divisor = pct[[col]]
            
    elif col[1].startswith('6 '):
        divisor = pct[[col]]
            
    elif col[1].startswith('7 '):
        divisor = pct[[col]]
            
    elif col[1].startswith('8 '):
        divisor = pct[[col]]
            
    elif col[1].startswith('9 '):
        divisor = pct[[col]]
        
    elif col[1].startswith('-1'):
        divisor = pct[[col]]
            
    pct.loc[:, [col]] = pct[[col]].values/divisor.values
```

```python
pct
```

```python
categories = ['0 Food and live animals ;', 
'1 Beverages and tobacco ;', 
'2 Crude materials, inedible, except fuels ;',
'3 Mineral fuels, lubricants and related materials ;', 
'4 Animal and vegetable oils, fats and waxes ;',
'5 Chemicals and related products, nes ;',
'6 Manufactured goods classified chiefly by material ;',
'7 Machinery and transport equipment ;',
'8 Miscellaneous manufactured articles ;', 
'9 Commodities and transactions not classified elsewhere in the SITC ;', 
'-1 TOTAL ;']

gross_pct = data.loc[:, pd.MultiIndex.from_product([['imports', 'exports'], categories])]

for col1, col2 in gross_pct.columns:
    
    gross_pct.loc[:, [(col1, col2)]] = gross_pct[[(col1, col2)]].values/gross_pct.loc[:, [(col1, '-1 TOTAL ;')]].values
```

```python
gross_pct
```

```python
gross_pct.columns = gross_pct.columns.to_flat_index()
```

## Change Column Names

```python
gross_pct.columns = [(col_1, col_2 + 'gross pct') for col_1, col_2 in gross_pct.columns]
pct.columns = [(col_1, col_2 + 'sub pct') for col_1, col_2 in pct.columns]
data.columns = [(col_1, col_2 + 'original data') for col_1, col_2 in data.columns]
ratio.columns = [(col_1, col_2 + 'ratio') for col_1, col_2 in ratio.columns]
```

## Create files

```python
data.index.name = 'Year'
pct.index.name = 'Year'
ratio.index.name = 'Year'
gross_pct.index.name = 'Year'
diff.index.name = 'Year'
```

```python
with pd.ExcelWriter('Data v4.xlsx') as writer:
    data.to_excel(writer, sheet_name = 'Original Data', index = True)
    diff.to_excel(writer, sheet_name = 'Difference', index = True)
    ratio.to_excel(writer, sheet_name = 'Ratio', index = True)
    pct.to_excel(writer, sheet_name = 'sub pct', index = True)
    gross_pct.to_excel(writer, sheet_name = 'gross_pct', index = True)
```
